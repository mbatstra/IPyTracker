```python
from matplotlib import pyplot as plt
from picamera2 import Picamera2
import numpy as np
import cv2

img = np.zeros((300, 300, 3), dtype=np.uint8); img[:, :, 0] = 255
plt.imshow(img)
plt.title('my picture')
plt.show()

cam = Picamera2()
cam.start()
img = cam.capture_array()
plt.imshow(img)
plt.title('my picture')
plt.show()
cam.close()
```

```python
from picamera2 import Picamera2, Preview
import time
camera = Picamera2()
camera_config = camera.create_preview_configuration()
camera.configure(camera_config)
camera.start_preview(Preview.DRM)
camera.start()
time.sleep(2)
raw = camera.capture_file()
camera.close()
```

```python
from picamera2 import Picamera2, Preview
from time import sleep
camera = Picamera2()
camera.start_preview()
camera.start()
sleep(10)
camera.stop_preview()
camera.close()
```

```python
from picamera2 import Picamera2
from matplotlib import pyplot as plt
from datetime import datetime
import numpy as np
import cv2

camera = Picamera2()
camera_config = camera.create_preview_configuration()
camera.configure(camera_config)

start = datetime.now()
delta = datetime.now() - start
print(delta.seconds)
while delta.seconds < 3:
    print('enters')
    img = camera.capture_array()
    print(img)
    plt.imshow(img)
    plt.title('my picture')
    plt.show()
    delta = datetime.now() - start
plt.imshow(img)
plt.title('my picture')
plt.show()
camera.close()
```
